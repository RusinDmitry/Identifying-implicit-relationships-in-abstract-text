```python
import pandas as pd
import numpy as np
clusters = pd.read_csv("/content/Identifying-implicit-relationships-in-abstract-text/clusters.csv")
```

```python
classes = {"Дата выхода": [57, 52],
 "Контингент (индификатор)": [33, 89, 58, 91, 75],
 "Цена": [49, 55, 56, 73, 85, 79, 53],
 "Название проекта (Описание)": [3, 5, 34, 7, 12, 19, 83, 76, 14, 61, 22, 25, 43, 48, 6, 41, 24, 95, 23, 6, 41, 43],
 "Номинальная мощность (кВт)": [27, 81],
 "Производитель, разработчик": [35, 13, 44, 74],
 "Покупатель, заказчик": [0, 42],
 "Участник торгов": [93]}
classes["Прочее"] = list(set(range(0, 100)) - set(sum(list(classes.values()), [])))


df_list = []
for index, row in clusters.iterrows():
    text_list = row.iloc[0]
    text = eval(text_list)
    class_ = np.full(shape = len(text), fill_value=row.iloc[1])
    df = pd.DataFrame({"text": text, "label": class_})
    df_list.append(df)


data = pd.concat(df_list, ignore_index=True)
```

```python
data2 = data.copy()
data2['label'] = data2['label'].astype(str)
for key, value in classes.items():
    for dig in value:
        data2['label'] = data2['label'].str.replace(r"\b"+str(dig)+r"\b", key, regex=True)

data2.to_csv("dataset.csv", index=False)
```
